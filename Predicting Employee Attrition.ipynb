# Data Types

Because I want to see the correlations for this entire dataset, everything has to be an integer or float to support the square root and other functions that go into pearsons' r value. At the very least, I need to change the Attrition column (the whole point of this study) to be able to be calculated in the correlations!

```python
types = ibm.dtypes
print(types)
```

```python
work = ibm.select_dtypes(include=[object])
print(work.head())
```

Looks like nine columns to convert. I'll map all of the subsequent responses to numerical values, then keep those values for any graphs later on (should they prove to be correlated). 

```python
#This will work on the Attrition and OverTime columns
yesno = {"Yes": 1,
        "No": 0}

#business travel
travel = {"Non-Travel": 0,
         "Travel_Rarely": 1,
         "Travel_Frequently": 2}

#department
dept = {"Sales": 0,
       "Research & Development": 1, 
        "Human Resources": 2}

#education field
education = {"Life Sciences": 0, 
            "Medical": 1,
            "Marketing": 2,
            "Technical Degree": 3,
            "Other": 4,
            "Human Resources": 5}

#gender
gender = {"Female": 1,
         "Male": 0}

#job role
job = {"Sales Executive":0,
       "Research Scientist": 1,
       "Laboratory Technician": 2,
       "Manufacturing Director": 3,
       "Healthcare Representative": 4,
       "Manager": 5,
       "Sales Representative": 6,
       "Research Director": 7,
        "Human Resources": 8}

#marital Status
married = {"Single": 0,
           "Married" :1,
           "Divorced" :2}

#over 18
adult = {"Y": 1,
        "N": 0}
```

```python
ibm["Attrition"].replace(yesno, inplace = True)
ibm["OverTime"].replace(yesno, inplace = True)
ibm["BusinessTravel"].replace(travel, inplace = True)
ibm["Department"].replace(dept, inplace = True)
ibm["EducationField"].replace(education, inplace = True)
ibm["Gender"].replace(gender, inplace = True)
ibm["JobRole"].replace(job, inplace = True)
ibm["MaritalStatus"].replace(married, inplace = True)
ibm["Over18"].replace(adult, inplace = True)
```

```python
ibm.head()
```

```python
ibm.dtypes
```

Now that everything is an interger, we can finally crank out some correlations.

# Correlations

```python
correlations = ibm.corr()
corr_a = correlations["Attrition"]
print(corr_a.sort_values())
```

Not a lot stands out besides Overtime. The general threshold is for a correlation value of .25 or more, Overtime barely makes the cut. It makes sense though, that there are a wide variety of factors that can contribute to something like this. Perhaps we're missing columns of key data, or perhaps the issue is that complex.

Remember, I also coded some of the options that could effect this outcome. Let's graph this and try to dig deeper.

# Graphing

```python
survey_fields = ibm.columns
survey_fields = [x for x in survey_fields]
survey_fields
```

```python
% matplotlib inline

plt.figure(figsize=(15,4))
pos = np.arange(len(corr_a))
plt.bar(pos, corr_a, align = 'center')
plt.xticks(pos, survey_fields, rotation = 80)
plt.title("Correlations between Survey Responses and Attrition")
plt.axhline(.15, c = "gray", alpha = 0.5)
plt.axhline(-.15, c = "gray", alpha = 0.5)
```

# Preliminary Results

I lowered the threshold to +/- .15 to capture more factors. Here, we have Age, Job Level, Monthly Income, Overtime, Total Working Years, Years with Current Role, and Years with Current Manager. Interestingly, besides Overtime, none of these are the ones that I coded earlier. It's worthwhile to note that while these have high correlations, we don't know which direction some of them lean. For example, age. It's correlated to attrition, but is that just for retired people or younger professionals getting their career going?

Some interesting items that were not correlated, Salary Hikes, Hourly Rates, Education, and Job Role. This is ficticous data, but those are all strange things to consider as not necessarily effecting attrition.
